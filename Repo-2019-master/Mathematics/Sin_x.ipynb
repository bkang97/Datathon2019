```python
import matplotlib.pyplot as plt
from math import sin,cos
import numpy as np
```

```python
ciclos=1000
```

```python
time=[sin(x) for x in np.linspace(0,3.14*20*ciclos,ciclos)]
plt.plot(time, c='r')
plt.ylim(-1,1)
plt.show()
```

```python
time=[sin(x)/2 for x in np.linspace(0,3.14*20*ciclos,ciclos)]
plt.plot(time, c='r')
plt.ylim(-1,1)
plt.show()
```

```python
time=[sin(x/2) for x in np.linspace(0,3.14*20*ciclos,ciclos)]
plt.plot(time, c='r')
plt.ylim(-1,1)
plt.show()
```

```python
time=[sin(x/2)/5+x/(ciclos*100) for x in np.linspace(0,3.14*20*ciclos,ciclos)]
plt.plot(time, c='r')
plt.show()
```

```python
time=[sin(x/2)/5+x/(ciclos*100) for x in np.linspace(0,3.14*20*ciclos,ciclos)]
time2=[cos(x/2)/5+x/(ciclos*100) for x in np.linspace(0,3.14*20*ciclos,ciclos)]
plt.plot(time, c='r')
plt.plot(time2, c='b')
plt.show()
```

```python
from scipy.fftpack import fft, ifft
import matplotlib.gridspec as gridspec
```

```python
period=5
steps=20
f = [fft(time[i:i+period]) for i in range(0,len(time)-period,steps)]
plt.plot(np.real(np.concatenate(f,axis=0)))
plt.show()
```

```python
plt.plot(np.imag(np.concatenate(f,axis=0)))
plt.show()
```
