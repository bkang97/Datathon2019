```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
energy = pd.read_csv('data/revenue_to_provider.csv')
```

```python
energy.head()
```

```python
energyTotal = energy[energy['industry_sector']=='Total Electric Industry']
energyTotal
```

```python
fig, ax = plt.subplots()

for key, grp in energyTotal.groupby(['state_code']):
    ax = grp.plot(ax=ax, kind='line', x='year', y='total', c=key, label=key)

plt.legend(loc='best')
plt.show()
```

```python
df = energyTotal.pivot(index='year', columns='state_code', values='total')
df
```
