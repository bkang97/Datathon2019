```python
import pandas as pd
import numpy as np
from collections import defaultdict
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
sedsData = pd.read_csv('datasets/seds.csv')
concentrationData = pd.read_csv('datasets/air_concentrations.csv')

concentrationData["date_off"] = pd.to_datetime(concentrationData['date_off'], format='%m/%d/%Y %H:%M:%S').dt.year
groupData = concentrationData.groupby(['state_code','date_off'])[['TSO4', 'TNH4', 'CA', 'MG','NA','K', 'CL','NSO4','NHNO3','WSO2','total_SO2','total_NO3']].mean()
ak_data = groupData.loc[('CT')]
ak_data["date"] = pd.Series([2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018], dtype='int').values
```
